색상의 대분류, 소분류 칼럼 모두에서 무의미한 값이 전체에서 큰 비율을 차지하지 않음.<br>
따라서 색상을 조금 더 직관적으로 확인할 수 있는 대분류 칼럼('perceived_colour_master_name')이 색상 분석에 더 용이할 듯.<br>
이 과정에서 'Unknown'같은 데이터가 적은 'value'는 제외한 dataframe으로 분석을 하게 되면 정확도가 더 높아질 듯.

#### 1-3 데이터가 적은 value를 제외한 데이터 프레임 구성

```python
#  perceived_colour_master_name'내 Unknown(685), undefined(105), Yellowish Green(5), Bluish Green(3)을 제거한 데이터 프레임 구축하기

# 제거하고 싶은 인덱스 구성
index_1 = art_data[(art_data['perceived_colour_master_name']=='Unknown') | 
                   (art_data['perceived_colour_master_name']=='undefined') |
                   (art_data['perceived_colour_master_name']=='Yellowish Green') |
                   (art_data['perceived_colour_master_name']=='Bluish Green')].index

print(index_1)
```

```python
# 인덱스에 해당하는 데이터 제외한 데이터프레임 구축
fin_art_colours=art_data.drop(index_1)
fin_art_colours
```
